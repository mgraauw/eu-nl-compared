```python
import pandas as pd
import os
```

```python
zibbase = 'zibs2020'
zibs = os.path.join(zibbase, 'mappable')
os.listdir(zibs)
```

```python
zib = 'Nl.zorg.Procedure-v5.2(2020EN).xlsx'
zibpath = os.path.join(zibs, zib)
zibdf = pd.read_excel(zibpath)
zibdf
```

```python
xlmbase = 'xtehr'
xlms = os.path.join(xlmbase, 'mappable')
os.listdir(xlms)
```

```python
xlm = 'Procedure.xlsx'
xlmpath = os.path.join(xlms, xlm)
xlmdf = pd.read_excel(xlmpath)
xlmdf
```

```python
mapbase = 'maps'
maps = os.path.join(mapbase, 'xtehr2zibs')
os.listdir(maps)
```

```python
file = 'Procedure'
mappath = os.path.join(maps, file + '.xlsx')
mapdf = pd.read_excel(mappath)
mapdf
```

```python
mapping = mapdf[['xtehr', 'zib']].merge(zibdf.add_suffix('_zib'), left_on='zib', right_on='path_zib', how='outer').merge(xlmdf.add_suffix('_xtehr'), left_on='xtehr', right_on='path_xtehr', how='outer')
mapping
```

```python
shortmap = mapping[['zib', 'xtehr', 'type_zib','type_xtehr', 'card._zib', 'card._xtehr']]
shortmap
```

```python
filepath = file + '-full.xlsx'
mappath = os.path.join(maps, filepath)
mapping.to_excel(mappath, index=False)
```

```python
filepath = file + '.csv'
mappath = os.path.join(maps, filepath)
mapping.to_csv(mappath, index=False)
```

```python
from IPython.display import Markdown
import numpy as np

content = "# " + mapping.iat[0, 0][4:] + "\n\n"
content += shortmap.fillna('').to_markdown(index=False) + "\n\n"
for i, row in mapping.iterrows():
    # print(i, row)
    content += "\n\n"
    if row.iloc[0] is not np.nan:
        content += "## " + str(row.iloc[0]) + "\n\n"
    else:
        content += "## zib: " + str(row.iloc[1]) + "\n\n"
    content += "### Table" + "\n\n"
    content += "| attribute | value |" + "\n" + "|---|---|" + "\n"
    for j, att in row.fillna('').items():
        content += "| " + str(j) + " | " + str(att) + " |" + "\n"
    content += "### Comments" + "\n\n"
with open(os.path.join(maps, file + '.md'), 'w') as f:
    f.write(content)
Markdown(content)
```
